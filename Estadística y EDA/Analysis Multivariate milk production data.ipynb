## Analysis of multivariate: milk production data

In this notebook we will use basic techniques for understanding multivariate relationships. For this analysis we will use Matplotlib and Seaborn libraries for data visualization and the basics pandas and numpy libraries.


The dataset is about milk production.

### Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
da = pd.read_csv("Bases/milknew.csv")
```

### Quantitative bivariate data

```python
da.head()
```

```python
sns.regplot(y="pH", x="Colour", data=da, fit_reg=False, scatter_kws={"alpha": 0.2})
```

```python
sns.jointplot(x="Temprature", y="pH", kind='kde', data=da)
```
